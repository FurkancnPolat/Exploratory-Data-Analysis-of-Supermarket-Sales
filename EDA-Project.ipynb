## Exploratory Data Analysis (EDA) in Sales


```python
# Load Data
import pandas as pd
df = pd.read_csv("supermarket_sales - Sheet1.csv")
```

```python
# Display the first few rows
df.head(10)
```

```python
# Check the dimensions of the dataset
print("Shape of the dataset:",df.shape)
```

```python
# Get information about the dataset
df.info()
# Check for missing values
df.isnull().sum()
```

```python
# Summary statistics for numerical columns
df.describe()
```

```python
# Check for duplicate rows
df.duplicated().sum()
# If duplicates exist, remove them
df.drop_duplicates(inplace=True)
```

```python
# For Sales distribution, we should import matplotlib and searborn
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,6))
sns.histplot(df["Total"], kde=True, bins=30)
plt.title("Distribution of Total Sales")
plt.xlabel("Total Sales")
plt.ylabel("Frequency")
plt.show()

```

```python
# Total sales by product line

plt.figure(figsize=(10,6))
sns.barplot(x="Product line", y="Total", data=df, estimator=sum)
plt.title("Total Sales by Product Line")
plt.xticks(rotation=45)
plt.show()
```

```python
# Total sales by gender
plt.figure(figsize=(8,5))
sns.barplot(x="Gender",y="Total",data=df,estimator=sum)
plt.title("Total Sales by Gender")
plt.show()
```

```python
# Time Series Analysis, For that We'll first convert the date column into a usable format.
# Convert the date column to datetime format
df["Date"]=pd.to_datetime(df["Date"])

# Extract day, month, and year for further analysis
df["Day"]= df["Date"].dt.day
df["Month"]= df["Date"].dt.month
df["Year"]= df["Date"].dt.year
```

```python
# Sales over time

plt.figure(figsize=(10,6))
df.groupby("Date")["Total"].sum().plot()
plt.title("Total Sales Over Time")
plt.ylabel("Total Sales")
plt.show()
```

```python
#Product Line Performance Over Time
# Group by product line and date to calculate monthly sales
monthly_sales = df.groupby([df['Date'].dt.to_period('M'), 'Product line'])['Total'].sum().unstack()

# Plot time series of sales by product line
monthly_sales.plot(figsize=(12,6))
plt.title('Monthly Sales by Product Line')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculate total quantity sold for each product line
top_products_by_quantity = df.groupby('Product line')['Quantity'].sum().reset_index()

# Sort by quantity sold in descending order
top_products_by_quantity = top_products_by_quantity.sort_values(by='Quantity', ascending=False)

# Display the results
print(top_products_by_quantity)
```

```python
# Calculate total sales by customer type
customer_type_sales = df.groupby('Customer type')['Total'].sum().reset_index()

# Sort the results by total sales in descending order
customer_type_sales = customer_type_sales.sort_values(by='Total', ascending=False)

# Display the results
print(customer_type_sales)
```

```python
# Calculate average transaction value by payment method
average_transaction_value = df.groupby('Payment')['Total'].mean().reset_index()

# Sort the results by average transaction value in descending order
average_transaction_value = average_transaction_value.sort_values(by='Total', ascending=False)

# Display the results
print(average_transaction_value)
```
